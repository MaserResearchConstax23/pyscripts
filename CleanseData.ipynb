```python
import pandas as pd
import math
import numpy as np
import astropy as ap

###############
## SETTINGS  ##
###############


search_rad = 1 #radial search radius in arcsec, deletes one entry of each pair in this radius

# Table names
RA_IDENT = 'ra'
DEC_IDENT = 'dec'
UID_IDENT = 'mcp_count'
# Path
DataPath = './data/AllSurveyed2021_NoDup_rudmingb.csv'
#Either Save Dupes or remove Dupes. Set To False to Remove dupes
SaveDupes = True 
#################


#read data, gives a DataFrame
all_surveyed = pd.read_csv(DataPath)

#################
##  Functions  ##
#################

def is_within_rad(ra_1, ra_2, dec_1, dec_2):
    return (search_rad / 3600.0) > (math.sqrt((math.pow((((ra_1 * (math.pi / 180.0)) - (ra_2 * (math.pi / 180.0))) * (math.cos((dec_1 * (math.pi / 180.0))))), 2)) + math.pow(((dec_1 * (math.pi / 180.0)) - (dec_2 * (math.pi / 180.0))), 2))) * (180.0 / math.pi)




#################
## Filter Data ##
#################


filtered_data = pd.DataFrame()


# go thru every entry in all_surveyed... if there is not already the same object 
# (by distance) in the filtered_data table, add the object to the filtered_data
for i, StartRow in all_surveyed.iterrows(): # i is numerical idx, StartRow is a Series
    is_dup = False
    for j, CompareRow in all_surveyed.iterrows():
        if StartRow[UID_IDENT] != CompareRow[UID_IDENT] and is_within_rad(StartRow[RA_IDENT], CompareRow[RA_IDENT], StartRow[DEC_IDENT], CompareRow[DEC_IDENT]):
            is_dup = True
    if (SaveDupes and is_dup) or (not SaveDupes and not is_dup):
        # insert the row into filtered_data
        filtered_data = filtered_data.append(StartRow, ignore_index=True)

print(filtered_data)
```
